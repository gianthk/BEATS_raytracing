## Read 3PW Spectral power files - XOP simulations

```python
# read CSV file with shadow results
names = ['E', 'P']
Spectral_power_gap11_file = '3PW_magnetic_field_profiles/Spectral_power_R06_g11.15.csv'
Spectral_power_gap15_file = '3PW_magnetic_field_profiles/Spectral_power_R06_g15.00.csv'
Spectral_power_gap20_file = '3PW_magnetic_field_profiles/Spectral_power_R06_g20.00.csv'
Spectral_power_gap30_file = '3PW_magnetic_field_profiles/Spectral_power_R06_g30.00.csv'
Spectral_power_gap40_file = '3PW_magnetic_field_profiles/Spectral_power_R06_g40.00.csv'

Spectral_power_gap11 = pd.read_csv(Spectral_power_gap11_file, sep=',', names=names, skiprows=[0])
Spectral_power_gap15 = pd.read_csv(Spectral_power_gap15_file, sep=',', names=names, skiprows=[0])
Spectral_power_gap20 = pd.read_csv(Spectral_power_gap20_file, sep=',', names=names, skiprows=[0])
Spectral_power_gap30 = pd.read_csv(Spectral_power_gap30_file, sep=',', names=names, skiprows=[0])
Spectral_power_gap40 = pd.read_csv(Spectral_power_gap40_file, sep=',', names=names, skiprows=[0])
# Spectral_power_gap11.info()
```

```python
# read CSV file with shadow results
names = ['E', 'F']
Flux_gap11_file = '3PW_magnetic_field_profiles/Flux_R06_g11.15.csv'
Flux_gap15_file = '3PW_magnetic_field_profiles/Flux_R06_g15.00.csv'
Flux_gap20_file = '3PW_magnetic_field_profiles/Flux_R06_g20.00.csv'
Flux_gap30_file = '3PW_magnetic_field_profiles/Flux_R06_g30.00.csv'
Flux_gap40_file = '3PW_magnetic_field_profiles/Flux_R06_g40.00.csv'

Flux_gap11 = pd.read_csv(Flux_gap11_file, sep=',', names=names, skiprows=[0])
Flux_gap15 = pd.read_csv(Flux_gap15_file, sep=',', names=names, skiprows=[0])
Flux_gap20 = pd.read_csv(Flux_gap20_file, sep=',', names=names, skiprows=[0])
Flux_gap30 = pd.read_csv(Flux_gap30_file, sep=',', names=names, skiprows=[0])
Flux_gap40 = pd.read_csv(Flux_gap40_file, sep=',', names=names, skiprows=[0])
# Flux_gap11.info()
```

```python
f, ax = plt.subplots()

plt.plot(Spectral_power_gap11['E'], Spectral_power_gap11['P'])
plt.plot(Spectral_power_gap15['E'], Spectral_power_gap15['P'])
plt.plot(Spectral_power_gap20['E'], Spectral_power_gap20['P'])
plt.plot(Spectral_power_gap30['E'], Spectral_power_gap30['P'])
plt.plot(Spectral_power_gap40['E'], Spectral_power_gap40['P'])
ax.set(xlim=(8e3, 1e5), ylim=(1e-8, 4e-2))

plt.legend(['gap: 11.15 mm', 'gap: 15.00 mm', 'gap: 20.00 mm', 'gap: 30.00 mm', 'gap: 40.00 mm'], loc='upper right', borderaxespad=0.2)

plt.grid(True, which="both")
ax.set_xscale('log')
ax.set_yscale('log')

plt.xlabel('Photon energy [eV]')
plt.ylabel('Spectral Power [W/eV]')
# f.savefig('Flux_tomo_BLs.png', bbox_inches='tight', dpi=600)
```
