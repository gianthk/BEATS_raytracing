## White beam flux in 1x1mm^2 @ sample - XOP power density simulations
Flux files for different combination of filters  are calculated in XOPPY and stored as .CSV in:
`/home/beatsbs/PycharmProjects/BEATS_raytracing/Oasys`

The photon flux is calculated with the following settings:
- primary slits OPEN
- 0.15 mm + 0.3 mm diamond windows thicknesses
- sample position (42m from photon source)
- through an aperture of 5mm x 5mm

```python
# read CSV file with Shadow results
names = ['Photon Energy [eV]', 'Flux [Photons/s/0.1%bw]']
# the last two columns are the intensity (or n. of good rays, I) and the Tot. rays (n)
FluxDensityWB_filecsv = '/home/beatsbs/PycharmProjects/BEATS_raytracing/Oasys/BEATS_BL_Flux_42m_5mmx5mm.csv'
FluxDensity20keV_filecsv = '/home/beatsbs/PycharmProjects/BEATS_raytracing/Oasys/BEATS_BL_Flux_42m_5mmx5mm_C5mm.csv'
FluxDensity36keV_filecsv = '/home/beatsbs/PycharmProjects/BEATS_raytracing/Oasys/BEATS_BL_Flux_42m_5mmx5mm_C5mm_Si5mm.csv'
FluxDensity69keV_filecsv = '/home/beatsbs/PycharmProjects/BEATS_raytracing/Oasys/BEATS_BL_Flux_42m_5mmx5mm_C5mm_Al0.2mm_W0.5mm.csv'

# FluxDensityWB = pd.read_csv(FluxDensityWB_filecsv, sep=',', names=names, skiprows=[0, 1, 2, 3, 4, 5])
FluxDensityWB = pd.read_csv(FluxDensityWB_filecsv, sep=',', names=names, skiprows=[0])
FluxDensity20keV = pd.read_csv(FluxDensity20keV_filecsv, sep=',', names=names, skiprows=[0])
FluxDensity36keV = pd.read_csv(FluxDensity36keV_filecsv, sep=',', names=names, skiprows=[0])
FluxDensity69keV = pd.read_csv(FluxDensity69keV_filecsv, sep=',', names=names, skiprows=[0])
```

Divide by the aperture area

```python
En = FluxDensityWB['Photon Energy [eV]']
FluxDensityWB_1mm = FluxDensityWB['Flux [Photons/s/0.1%bw]'][:]/25
FluxDensity20keV_1mm = FluxDensity20keV['Flux [Photons/s/0.1%bw]'][:]/25
FluxDensity36keV_1mm = FluxDensity36keV['Flux [Photons/s/0.1%bw]'][:]/25
FluxDensity69keV_1mm = FluxDensity69keV['Flux [Photons/s/0.1%bw]'][:]/25
```

## Pink beam with filters - XOP and Shadow simulations

Transmissivity files were calculated with the Power widget in XOP and for different filter settings:
- Al [0.1, 0.2, 0.5, 1.0] [mm]
- Cu [10, 20, 50] [um]
- Fe [10, 50] [um]
- Pd [75 um]
- Graphite [5, 20] [mm]
- Diamond [0.7 mm]
- Air [1 m]

```python
filters=['E', '100umAl', '200umAl', '500umAl', '1mmAl', '10umCu', '20umCu', '50umCu', '10umFe', '50umFe', '75umPd', '5mmGraphite', '20mmGraphite', '700umCVD', '1mAir', '1.5mmSi', '25umMo']
```

```python
# read CSV trasnmissivity file
transmissivity = pd.read_csv('Oasys/transmissivity_BEATS_filters.csv', sep=',', names=filters, skiprows=[0])
```

```python
# Sample transmissivity data @ the energies used for the Shadow simulation
T_data = np.zeros((18,17),dtype=np.float)
for i in range(0, 18):
    T_data[i,:] = transmissivity.loc[np.argmin(np.abs(float(FluxDensityWB['E'][i])-transmissivity['E']))]
    
T_data2 = np.zeros((10,17),dtype=np.float)
for i in range(0, 10):
    T_data2[i,:] = transmissivity.loc[np.argmin(np.abs(float(FluxDensityWB['E'][38+i])-transmissivity['E']))]
```

```python
# check plot of the transmissivity of one filter
f, ax = plt.subplots()
plt.plot(1e-3*transmissivity['E'][200:], transmissivity['75umPd'][200:])
plt.plot(1e-3*T_data[10:,0], T_data[10:,10], 'or')
ax.set_xscale('log')
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.set_yscale('log')
plt.xlabel('PhotonEnergy [keV]')
plt.ylabel('Transmissivity $75 \mu m Pd$ [%]')
plt.grid(True, which="both")
```
