## 07f_b: White beam size at sample

### Primary slits OPEN (effect of the fix mask)

#### Fix mask size

```python
accept_angle = 1e-3 # [rad]
d = 6500e-3 # [m] distance fixed mask - source
print("Fix mask Hor. size for %.2f mrad acceptance angle: %.3f mm"%(accept_angle*1e3, 1e3*accept_angle*d))
```

#### FOV with foreseen with 1x magnification microscope and PCO.edge5.5 detector
See values for the microscope optics @ ID19 [here](http://www.esrf.eu/UsersAndScience/Experiments/Imaging/ID19/BeamlineDescription/Detectors/Frelon_intro/Frelon_optics)

```python
PCOedge_pixels = 2560
mic_pixelsize = np.array([0.72, 3.61, 7.2, 0.33, 3.25, 6.6]) # [micron] (Max. and Min.) OptiquePeter TwinMic and Revolver microscopes
FOV = 1e-3*PCOedge_pixels*mic_pixelsize # [mm]
print("Maximum Horizontal FOV: %.3f mm"%(FOV.max()))
```

#### Photon source properties from Shadow (8 - 50 keV) 

```python
ss_x = 1839e-6 # [m] Hor. photon source size (FWHM)
ss_y = 32e-6 #  [m] Ver. photon source size (FWHM)
ss_x_p = 8210e-6 # [rad] Hor. divergence (FWHM)
ss_y_p = 237e-6 # [rad] Ver. divergence (FWHM)
```

#### Geometrical props. of the beamline

```python
a2 = 12.994e-3 # [m] fixed mask aperture (2 mrad)
a15 = 9.745e-3 # [m] fixed mask aperture (1.5 mrad)
a1 = 6.497e-3 # [m] fixed mask aperture (1 mrad)
dw = 32075e-3 # [m] distance CVD - source
```

#### Effective divergence

```python
eta_x2 = 2*np.sqrt((ss_x/2)**2 + (a2/2)**2)/(d)
eta_x15 = 2*np.sqrt((ss_x/2)**2 + (a15/2)**2)/(d)
eta_x1 = 2*np.sqrt((ss_x/2)**2 + (a1/2)**2)/(d)
eta_y = 2*np.sqrt((ss_y/2)**2 + (a1/2)**2)/(d)
print(f"Effective Hor. divergence: {accept_angle*1e3} -> %.4f [mrad]"%((eta_x1*1e3)))
```

#### Beam size @ sample

```python
# HORIZONTAL
SDD = np.arange(31.8, 45.2, 0.2) # Sample Detector Distance [m]
f = plt.figure()
plt.plot([dw, dw], [34, 70], '--r')
plt.plot(SDD, 1e3*(ss_x+(np.sin(eta_x2)*SDD)), 'k')
plt.plot(SDD, 1e3*(ss_x+(np.sin(eta_x15)*SDD)), '-.k')
plt.plot(SDD, 1e3*(ss_x+(np.sin(eta_x1)*SDD)), '--k')
plt.xlabel('Distance from source [m]')
plt.ylabel('Hor. Beam size @ sample [mm]')
plt.xticks(np.arange(32, 46, 1))
plt.yticks(np.arange(30, 95, 5))
plt.grid(True, which="both")
plt.legend(['CVD', '2 mrad', '1.5 mrad', '1 mrad'], loc='upper left')
# f.savefig("H_FOV_1-2mrad.png", bbox_inches='tight', dpi=600)

# VERTICAL
f = plt.figure()
plt.plot(SDD, 1e3*(ss_y+(np.sin(ss_y_p)*SDD)), 'k')
plt.xlabel('Distance from source [m]')
plt.ylabel('Ver. Beam size @ sample [mm]')
plt.xticks(np.arange(32, 46, 1))
plt.grid(True, which="both")
# f.savefig("V_FOV.png", bbox_inches='tight', dpi=600)
```

### Primary slits as secondary source

#### Import Shadow data

```python
names = ['slit', 'FWHM_X', 'FWHM_Z', 'tot_rays', 'good_rays', 'FWHM_Xp']
beamsizefilecsv = 'BEATS_BL_beam_size.csv'
beamsizedata = pd.read_csv(beamsizefilecsv, sep=',', names=names, skiprows=[0, 1, 2, 3, 4])
```

#### Effective horizontal divergence

```python
a = np.arange(2.0e-3, 0.1e-3, -0.05e-3) # primary slit aperture [m]
d_slits = 7.635 # [m] distance source - slits
d_sample = 43 # [m]

eta_x_2 = np.sqrt((ss_x/2)**2 + (a/2)**2)/(d_slits)   # half divergence
```

```python
# Plot and compare with Shadow simulations
f = plt.figure()
plt.plot(a*1e3, 2e3*eta_x_2, '--k')
plt.plot(beamsizedata['slit'][1:9].astype('float'), 1e-3*beamsizedata['FWHM_Xp'][1:9].astype('float'), '-ok')
plt.xticks(np.arange(0.2, 2.1, 0.2))
plt.xlabel('Primary slits aperture [mm]')
plt.ylabel('Effective Hor. divergence [mrad] (FWHM)')
plt.grid(True, which="both")
plt.legend(['$sqrt((\sigma_x/2)^2+(a/2)^2)/d$', 'Shadow simulation (8-50 keV)', '10 keV', '20 keV', '40 keV'], loc='upper left')
f.savefig("slit_VS_H_beam_divergence.png", bbox_inches='tight', dpi=600)
```

#### Horizontal beam size @ 43 m

```python
f, ax = plt.subplots()
plt.plot(a*1e3, 1e3*((2*eta_x_2)*(d_sample-d_slits)), '--k')
plt.plot(beamsizedata['slit'][1:9].astype('float'), 1e-3*beamsizedata['FWHM_X'][1:9].astype('float'), '-ok')
plt.xticks(np.arange(0.2, 2.1, 0.2))

plt.xlabel('Primary slits aperture [mm]')
plt.ylabel('Hor. beam size @ 43m [mm] (FWHM)')
plt.grid(True, which="both")
plt.legend(['$2\eta_x d$', 'Shadow simulation (8-50 keV)', '10 keV', '20 keV', '40 keV'], loc='upper left')

f.savefig("slits_VS_Hor_beam_size.png", bbox_inches='tight', dpi=600)
```

# 07f_c: Coherence length and power
For the calculation of spatial coherence length and image blur see:

Cloetens, Peter. 2007. “Phase Contrast Imaging - Coherent Beams.” presented at the School on X-ray Imaging Techniques at the ESRF, ESRF, Grenoble, France, February 5. http://www.esrf.eu/files/live/sites/www/files/events/conferences/2007/xray-imaging-school/Presentations/07_Cloetens.pdf.


### Primary slits aperture VS coherence length

```python
eN = np.arange(10e3,30e3,10e3) # 10 - 20 keV
l = (h*c)/(eN*1.6e-19) # wavelength [m]
ss_x_0 = 1.856e-3 # [m] Hor. photon source size (FWHM) @ 20keV
ss_x = np.arange(1.0e-3, 0.3e-3, -0.01e-3) # secondary source size (slit aperture) [m]
ss_x_topotomo = [0.5e-3, 0.5e-3] # [m] (FWHM)
z = 33 # [m] distance primary slits - sample @ 41m - up to 35 m actually
z_topotomo = [33, 33] # [m]
lc_topotomo = (np.multiply(l,z_topotomo)/np.multiply(2,ss_x_topotomo))*1e6

print("Coherence length @ 20keV:")
print("- ID19@ESRF (l=145m): %.2f [micron]"%((l[1]*145/(2*25e-6))*1e6))
print("- TOMCAT@SLS (l=34m): %.2f [micron]"%((l[1]*34/(2*140e-6))*1e6))
print("- SYRMEP@Elettra (l=23m): %.2f [micron]"%((l[1]*23/(2*197e-6))*1e6))
print("- TopoTomo@ANKA (l=33m): %.2f [micron]"%((l[1]*33/(2*0.5e-3))*1e6))
print("- BEATS - Primary slits OPEN (l=43m): %.2f [micron]"%((l[1]*43/(2*ss_x_0))*1e6))
```

```python
# plots
f = plt.figure()
plt.plot(ss_x*1e3,(l[0]*z/(2*ss_x))*1e6, 'k') # 10keV
plt.plot(ss_x*1e3,(l[1]*z/(2*ss_x))*1e6, '--k') # 20keV
plt.plot(np.multiply(ss_x_topotomo, 1e3), lc_topotomo, 'or') # TopoTomo@ANKA

ax.set(xlim=(0.3, 1.1), ylim=(0.5, 7))
plt.xticks(np.arange(0.3, 1.1, 0.1))
plt.yticks(np.arange(0.5, 7, 0.5))
plt.grid(True, which="both")
plt.legend(['10 keV', '20 keV', 'TopoTomo@ANKA'])

plt.xlabel('Primary slits aperture [mm]')
plt.ylabel('Hor. coherence length [$\mu m$]')
# f.savefig("slits_VS_Hor_coherence_length.png", bbox_inches='tight', dpi=600)
```

### Slit aperture VS image blur

```python
z1 = 32 # [m]
z2 = [0.1, 0.2, 0.4] # [m]
ss_x_topotomo = [0.5e-3, 0.5e-3, 0.5e-3] # [mm] (FWHM)

# plots
f = plt.figure()
plt.plot(ss_x*1e3, (z2[0]/z1)*ss_x*1e6, 'k') # SDD 10 cm
plt.plot(ss_x*1e3, (z2[1]/z1)*ss_x*1e6, '--k') # SDD 20 cm
plt.plot(ss_x*1e3, (z2[2]/z1)*ss_x*1e6, '-.k') # SDD 40 cm
plt.plot(np.multiply(ss_x_topotomo, 1e3), np.divide(z2, z1)*0.5*1e3, 'or') # TopoTomo@ANKA
plt.yticks(np.arange(1, 13, 1))

plt.grid(True, which="both")
plt.legend(['SDD 10 cm', 'SDD 20 cm', 'SDD 40 cm', 'TopoTomo@ANKA'])

plt.xlabel('Primary slits aperture [mm]')
plt.ylabel('Hor. blur [$\mu m$]')
f.savefig("slits_VS_Hor_blur.png", bbox_inches='tight', dpi=600)
```
